```python
from google.colab import drive
drive.mount('/content/drive')

from os import listdir
from matplotlib import image
from PIL import Image

loaded_images = list()
for filename in listdir('/content/drive/MyDrive/grayscale/folders'):
  for img in listdir('/content/drive/MyDrive/grayscale/folders/' + filename):
    img_data = image.imread('/content/drive/MyDrive/grayscale/folders/'+ filename + '/' + img)
    loaded_images.append(img_data)
    print('> loaded %s %s' % (filename, img_data.shape))

def resize():
    files = listdir('/content/drive/MyDrive/grayscale/folders/')
    for item in files:
            image = Image.open(item)
            image.thumbnail((64,64))
            print(image.size)

resize()
```
